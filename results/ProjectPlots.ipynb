#R

##1 thread

```python
import re
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = '/content/drive/My Drive/Project/R_15_100_1thread.txt'

# Initialize a list to store parsed data
data = []

# Define a regex pattern to extract R, function type and the output
pattern = r"R=(\d+) vamana_function=(\w+): ([\d.]+),([\d.]+),([\d.]+),([\d.]+),([\d.]+),([\d.]+),([\d.]+)"

with open(file_path, 'r') as file:
    for line in file:
        match = re.search(pattern, line)
        if match:
            R = int(match.group(1))
            function_type = match.group(2)
            values = list(map(float, match.groups()[2:9]))
            data.append([R, function_type, *values])

# Create a DataFrame
columns = [
    "R", "Function", "Recall@100", "Recall@100 for Unfiltered",
    "Recall@100 for Filtered", "Queries per second",
    "Average Search Time", "Indexing time", "Graph size"
]
dfR = pd.DataFrame(data, columns=columns)

# Separate tables for each vamana_function
vamana_tables = {}
for function in dfR['Function'].unique():
    vamana_tables[function] = dfR[dfR['Function'] == function]

# Create plots for each function with subplots for the 7 values
for function in dfR['Function'].unique():
    filtered_dfR = dfR[dfR['Function'] == function]
    n_plots = len(columns) - 2  # Number of columns to plot (excluding 'R' and 'Function')
    fig, axes = plt.subplots((n_plots + 1) // 2, 2, figsize=(12, 14))  # Rows = ceil(n_plots/2), 2 columns
    axes = axes.flatten()
    fig.suptitle(f'Plots for {function.capitalize()}', fontsize=16)

    for i in range(n_plots):
        ax = axes[i]
        ax.plot(filtered_dfR['R'], filtered_dfR.iloc[:, i + 2], marker='o', label=columns[i + 2])
        ax.set_title(columns[i + 2])
        ax.set_xlabel('R')
        ax.set_ylabel(columns[i + 2])
        ax.legend()
        ax.grid()

    # Hide unused subplots
    for j in range(n_plots, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to include the title
    plt.show()
```
